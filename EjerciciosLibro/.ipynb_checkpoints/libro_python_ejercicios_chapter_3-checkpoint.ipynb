```python
'''
3.1 (Validating User Input) Modify the script of Fig. 3.3 to validate its inputs. For any
input, if the value entered is other than 1 or 2, keep looping until the user enters a correct value. 
Use one counter to keep track of the number of passes, then calculate the number
of failures after all the user’s inputs have been received.
'''
```

```python
# initialize the variables
passes = 0 #number of passes
failures = 0 #number of failures

#process 10 students
for student in range(10):
    #get one exam result
    result = int(input('Enter result (1=pass, 2=fail): '))
    
    while result != 1 and result != 2: #condicion que debe ser true para mantener el loop
        print("Invalid value, please type either 1 or 2") #mensaje para comunicar que es lo que debe cambiar
        result = int(input('Enter result (1=pass, 2=fail): ')) #la variable debe cambiar para poder romper el loop, sin esto seria un loop infinito
    
    if result == 1:
        passes = passes + 1
    else:
        failures = failures + 1
        
#termination phase
print('Passed:', passes)
print('Failed:', failures)

if passes > 8:
    print('Bonus to instructor')
```

```python
'''
3.5 (if…else Statements) Reimplement the script of Fig. 2.1 using three if…else
statements rather than six if statements. [Hint: For example, think of == and != as “opposite”
tests.]
'''

print('Enter two integers, and I will tell you the relationships they satisfy.')

#read the first integer
num1= int(input("Enter the first integer:"))

#read the second integer
num2= int(input("Enter the second integer:"))

#three if...else statements!

#evaluate equal/not equal
if num1 == num2:
    print(num1, "is equal to", num2)
else:
    print(num1, "is not equal to", num2)
    
#evaluate less/greater
if num1 > num2:
    print(num1, "is greater than", num2)
else:
    print(num1, "is less than", num2)

#evaluate less than or equal to/greater than or equal to
if num1 <= num2:
    print(num1, "is less than or equal to", num2)
else:
    print(num1, "is greater than or equal to", num2)


```

```python
'''
3.7 (Table of Bacteria) In Exercise 2.8, you wrote a script to calculate the growth in
the number of bacteria after 0, 5, 10 and 15 hours and printed the resulting values in table
format. Reimplement your script using a for loop and the f-string capabilities you learned in this chapter to produce the following table 
with the numbers right aligned in each column.

'''
```

```python
def calculate_bacteria(n_hours):
    hours = str(n_hours)
    bacteria = str(200 * 2 ** n_hours) 
    tab = "\t"
    
    print(f'{hours:>4} {tab} {bacteria:>19}')
    
    
list_hours = [0, 5, 10, 15]
    
print("Hour \t  Number of Bacteria" )

for x in list_hours:
    calculate_bacteria(x)
```

```python
'''
3.8 (Flu Patients Data Processing) During the flu season, it is important to keep track
of the number of infected patients. Write a script in which the user can input the number
of reported infections per day over one week. Calculate the total, average, smallest, and
largest of these values. Write your script using a loop structure.
'''

#hay que inicializar la lista vacia primero
list_of_infected = []

for x in range(7):
    list_of_infected.append(int(input("Type the number of infected patients on Day #" + str(x + 1) + ":" ))) #añadir los valores con append

#total
total = sum(list_of_infected)
print("the total is " + str(total))

#average
amount_of_values = len(list_of_infected)
average = total / amount_of_values
print("the average is " + str(average))

#smallest value
smallest = min(list_of_infected)
print("the smallest value is " + str(smallest))

#largest value
largest = max(list_of_infected)
print("the largest value is " + str(largest))
```

```python
'''
3.9 (Calculating Time) In Exercise 2.11, you wrote a script that calculated the number
of hours, minutes, and remaining seconds based on the number of seconds received
through user input. Reimplement your script to use a loop that in an iterative process
“picks off” the hours, minutes, and remaining seconds using the // and % operators.
'''

while True:
    try:
        # Ask the user for the number of seconds
        seconds = int(input("Please enter the number of seconds: "))
        
        # Convert the seconds to hours, minutes and remaining seconds
        hours = seconds // 3600
        seconds %= 3600
        minutes = seconds // 60
        seconds %= 60
        
        # Print the result
        print(f"{hours} hour(s), {minutes} minute(s) and {seconds} second(s)")
        
        # Ask the user if they want to continue
        cont = input("Do you want to continue? (yes/no): ")
        if cont.lower() != "yes":
            break
            
    except ValueError:
        print("Invalid input. Please enter a valid number.")

```

```python
'''
3.10 (Hourly Wage Calculator) Reimplement Exercise 2.12 to use a loop that calculates
and displays the hourly wage an employee earns after a good performance review for years
1 through 10.
'''

'''
Use the following formula to calculate these wages:
w = o(1 + p)n
where
w is the new hourly wage,
o is the original hourly wage,
p is the percentage increase or decrease, and
n is the number of years with an increase or decrease in hourly wage.
'''

#variables base
o = 10
p = 0.03
n = 0

#loop de años buenos durante 1 año... hasta 10 años
while(n <= 10):
    n += 1
    good_year= o * ((1 + p) ** n)
    print("Only Good reviews for year #{} has a wage of: ${:,.2f}".format(n, good_year))
    
```

```python
# initialization
amt_babies_so_far=0
amt_does_so_far=0

#processing phase
does_n = int(input("enter the number of does in the rabbit colony (-1 to end):"))

while(does_n != -1):
    
    #zero division validation
    if (does_n == 0):
        print("can't divide by zero!!!!!!")
        does_n = int(input("enter the number of does in the rabbit colony (-1 to end):"))
    else:
        rabbits_born_n = int(input("enter the number of baby rabbits born in the past month:"))
        avg = rabbits_born_n / does_n
        amt_babies_so_far += rabbits_born_n
        amt_does_so_far += does_n
        avg_so_far= amt_babies_so_far / amt_does_so_far
        
        print("On average {} baby rabbits were born for each doe".format(avg))
    
        does_n = int(input("enter the number of does in the rabbit colony (-1 to end):"))
    
        #termination phase
        if (does_n == -1):
            print("Total number of baby rabbits for each doe so far: {}".format(avg_so_far))
    
```

```python
'''
3.13 (Perfect Numbers) In number theory, a perfect number is a positive integer that is
equal to the sum of its divisors. Perfect numbers were first studied by the Pythagoreans
who thought that they had mystical properties. They were also extensively studied by
Greeks (including Euclid) for their numerological properties.
The smallest perfect number is 6, because 6 = 3 + 2 + 1, with 3, 2, and 1 being the divisors.
Other examples of perfect numbers are: 28, 496 and 8128. Write a script that inputs
a nonnegative integer and displays whether it is a perfect number or not.
'''

#ask for the number
num = int(input("enter the number to check if it is a perfect number:"))
divisors = []

#get the divisors in a list
print("the divisors of the number {} are :".format( num))
for x in range(1,num): #(this range starts at 1 and ends with the num since perfect numbers always exclude themselves from the divisor list)
    if(num%x == 0):
        print(x)
        divisors.append(x)
        
#sum all the items on the list
sum_all_divisors = sum(divisors)

#compare if it is the same number as the previously inputted
print("the sum of all divisors is = " + str(sum_all_divisors))

if(sum_all_divisors == num):
    print("this means this number is indeed, a perfect number! :)")
else:
    print("this means this number is not a perfect number :(")
```
