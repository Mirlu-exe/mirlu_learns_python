```python
'''
3.1 (Validating User Input) Modify the script of Fig. 3.3 to validate its inputs. For any
input, if the value entered is other than 1 or 2, keep looping until the user enters a correct value. 
Use one counter to keep track of the number of passes, then calculate the number
of failures after all the user’s inputs have been received.
'''
```

```python
# initialize the variables
passes = 0 #number of passes
failures = 0 #number of failures

#process 10 students
for student in range(10):
    #get one exam result
    result = int(input('Enter result (1=pass, 2=fail): '))
    
    while result != 1 and result != 2: #condicion que debe ser true para mantener el loop
        print("Invalid value, please type either 1 or 2") #mensaje para comunicar que es lo que debe cambiar
        result = int(input('Enter result (1=pass, 2=fail): ')) #la variable debe cambiar para poder romper el loop, sin esto seria un loop infinito
    
    if result == 1:
        passes = passes + 1
    else:
        failures = failures + 1
        
#termination phase
print('Passed:', passes)
print('Failed:', failures)

if passes > 8:
    print('Bonus to instructor')
```

```python
'''
3.5 (if…else Statements) Reimplement the script of Fig. 2.1 using three if…else
statements rather than six if statements. [Hint: For example, think of == and != as “opposite”
tests.]
'''

print('Enter two integers, and I will tell you the relationships they satisfy.')

#read the first integer
num1= int(input("Enter the first integer:"))

#read the second integer
num2= int(input("Enter the second integer:"))

#three if...else statements!

#evaluate equal/not equal
if num1 == num2:
    print(num1, "is equal to", num2)
else:
    print(num1, "is not equal to", num2)
    
#evaluate less/greater
if num1 > num2:
    print(num1, "is greater than", num2)
else:
    print(num1, "is less than", num2)

#evaluate less than or equal to/greater than or equal to
if num1 <= num2:
    print(num1, "is less than or equal to", num2)
else:
    print(num1, "is greater than or equal to", num2)


```

```python
'''
3.7 (Table of Bacteria) In Exercise 2.8, you wrote a script to calculate the growth in
the number of bacteria after 0, 5, 10 and 15 hours and printed the resulting values in table
format. Reimplement your script using a for loop and the f-string capabilities you learned in this chapter to produce the following table 
with the numbers right aligned in each column.

'''
```

```python
def calculate_bacteria(n_hours):
    hours = str(n_hours)
    bacteria = str(200 * 2 ** n_hours) 
    tab = "\t"
    
    print(f'{hours:>4} {tab} {bacteria:>19}')
    
    
list_hours = [0, 5, 10, 15]
    
print("Hour \t  Number of Bacteria" )

for x in list_hours:
    calculate_bacteria(x)
```

```python
'''
3.8 (Flu Patients Data Processing) During the flu season, it is important to keep track
of the number of infected patients. Write a script in which the user can input the number
of reported infections per day over one week. Calculate the total, average, smallest, and
largest of these values. Write your script using a loop structure.
'''

#hay que inicializar la lista vacia primero
list_of_infected = []

for x in range(7):
    list_of_infected.append(int(input("Type the number of infected patients on Day #" + str(x + 1) + ":" ))) #añadir los valores con append

#total
total = sum(list_of_infected)
print("the total is " + str(total))

#average
amount_of_values = len(list_of_infected)
average = total / amount_of_values
print("the average is " + str(average))

#smallest value
smallest = min(list_of_infected)
print("the smallest value is " + str(smallest))

#largest value
largest = max(list_of_infected)
print("the largest value is " + str(largest))
```

```python
'''
3.9 (Calculating Time) In Exercise 2.11, you wrote a script that calculated the number
of hours, minutes, and remaining seconds based on the number of seconds received
through user input. Reimplement your script to use a loop that in an iterative process
“picks off” the hours, minutes, and remaining seconds using the // and % operators.
'''

while True:
    try:
        # Ask the user for the number of seconds
        seconds = int(input("Please enter the number of seconds: "))
        
        # Convert the seconds to hours, minutes and remaining seconds
        hours = seconds // 3600
        seconds %= 3600
        minutes = seconds // 60
        seconds %= 60
        
        # Print the result
        print(f"{hours} hour(s), {minutes} minute(s) and {seconds} second(s)")
        
        # Ask the user if they want to continue
        cont = input("Do you want to continue? (yes/no): ")
        if cont.lower() != "yes":
            break
            
    except ValueError:
        print("Invalid input. Please enter a valid number.")

```
