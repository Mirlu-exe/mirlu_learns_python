```python

%matplotlib inline
import matplotlib.pyplot as plt 

def generate_bar_chart(labels, values):
    fig, ax = plt.subplots()
    ax.bar(labels, values)
    
    plt.show()
    
def generate_pie_chart(labels, values):
    fig, ax = plt.subplots()
    ax.pie(values, labels = labels) #los labels hay que especificarlos asi, es una regla de la libreria que hay que seguir 
    ax.axis('equal')
    plt.show
    
if __name__ == '__main__':
    
    labels = ['a', 'b', 'c']
    values = [10,40,800]
    
    generate_bar_chart(labels, values)
    
    generate_pie_chart(labels, values)
        
```
