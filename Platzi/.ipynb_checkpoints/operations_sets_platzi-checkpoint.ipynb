```python
''' Union '''
```

```python
set_a = {'col', 'mex', 'bol'}
set_b = {'pe', 'bol'}
set_c = set_a.union(set_b)
print(set_c)
```

```python
#tambien puedes usar un operador en vez de un metodo, el resultado es igual, solo es una sintaxis diferente.
print(set_a | set_b)
```

```python
''' Intersection '''
```

```python
set_c = set_a.intersection(set_b)
print(set_c)
```

```python
#igual con el operando
print(set_a & set_b)
```

```python
''' Difference '''
```

```python
set_c = set_a.difference(set_b)
print(set_c)
```

```python

#con operador seria:
print(set_a - set_b)
```

```python
''' Symmetric Difference '''
```

```python
set_c = set_a.symmetric_difference(set_b)
print(set_c)
```

```python
#con el operando seria:
print(set_a ^ set_b)
```
